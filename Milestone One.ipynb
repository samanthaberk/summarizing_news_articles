### Step 2: Segregate the news and the respective summaries reflected in the dataset

```python
import glob
```

```python
# Initialize an empty list to store the segregated summaries and their corresponding news. 
master_list = []
```

```python
# Segregate each story into actual news and their corresponding summaries. 
# Remove the whitespaces around each summary
# Maintain a dictionary to store the list of different summaries along with their corresponding news. 
```

```python
# investigate the contents of the first story
f = open('638ba1352bdf405a8f5bd681d7fe5c928686afff.story', 'r')
for line in f:
    print(line)
```

```python
def get_summaries(file_list):
    list = []
    for i in range(1, len(file_list)):
        list.append(file_list[i].strip())
    return list

# process each file and add them as segemented dictionaries to the master list
for file in glob.glob(os.getcwd() + '/*'):
    with open(file) as f:
        text = f.read()
        f = " ".join(text.splitlines()) #get rid of new lines
        file_list = f.split('@highlight') #get list segmented on @highlight
    
        # store news story as the key and summaries as a list of values
        dict = {}
        dict[file_list[0]] = get_summaries(file_list)
    
        # add dict to master_list
        master_list.append(dict)
        

# with open('638ba1352bdf405a8f5bd681d7fe5c928686afff.story') as file:
#     text = file.read()
#     f = " ".join(text.splitlines())
#     file_list = f.split('@highlight')
#     print(len(file_list))
```
